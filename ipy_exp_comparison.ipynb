```python
from models import CSKN8, CRFFNet8
from exp_comparison_DKLModel import DenseNetFeatureExtractor, DKLModel
from exp_comparison_PretrainedModels import PretrainModel
import matplotlib.pyplot as plt
import torch
import gpytorch
from thop import profile

model_arr = ['CSKN8', 'CRFFNet8', 'DKL', 'resnet', 'densenet', 'vgg', 'shufflenet']
flops_arr = []
params_arr = []

lr0 = 1e-4
lr1 = 1e-5
sigma0 = 1e-6
sigma1 = 1e-6
growth_factor = 4
input = torch.randn(1, 1, 28, 28)
flops, params = profile(CSKN8(512, 10, 1, sigma0, sigma1, growth_factor), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)

flops, params = profile(CRFFNet8(512, 10, 1, sigma0), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)

input = torch.randn(1, 1, 32, 32)
feature_extractor = DenseNetFeatureExtractor(in_channels=1, block_config=(6, 6, 6), num_classes=10)
num_features = feature_extractor.classifier.in_features
model = DKLModel(feature_extractor, num_dim=num_features)
likelihood = gpytorch.likelihoods.SoftmaxLikelihood(num_features=model.num_dim, num_classes=10)
flops, params = profile(model, inputs = (input, ))
output = model(input)
flops1, params1 = profile(likelihood, inputs = (output, ))
print('DKL', flops, params)
flops_arr.append(flops + flops1)
params_arr.append(params + params1)

network = 'resnet'
flops, params = profile(PretrainModel(network, 1, 10), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)

network = 'densenet'
flops, params = profile(PretrainModel(network, 1, 10), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)
print(network, flops, params)

network = 'vgg'
flops, params = profile(PretrainModel(network, 1, 10), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)
print(network, flops, params)

network = 'shufflenet'
flops, params = profile(PretrainModel(network, 1, 10), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)

for i in range(len(params_arr)):
    print('{} --- FLOPs: {:.2f} M, # Params: {:.2f} M'.format(model_arr[i], flops_arr[i]/1000**2, params_arr[i]/1000**2))
```
