```python
from models import CSKN8, CRFFNet8, PretrainModel
import matplotlib.pyplot as plt
import torch
from thop import profile

model_arr = ['CSKN8', 'CRFFNet8', 'resnet', 'vgg', 'densenet', 'shufflenet']
flops_arr = []
params_arr = []

lr0 = 1e-4
lr1 = 1e-5
sigma0 = 1e-6
sigma1 = 1e-6
growth_factor = 4
input = torch.randn(1, 1, 32, 32)
flops, params = profile(CSKN8(2048, 10, 1, sigma0, sigma1, growth_factor), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)

flops, params = profile(CRFFNet8(2048, 10, 1, sigma0), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)

network = 'resnet'
flops, params = profile(PretrainModel(network, 1, 10), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)

network = 'vgg'
flops, params = profile(PretrainModel(network, 1, 10), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)

network = 'densenet'
flops, params = profile(PretrainModel(network, 1, 10), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)

network = 'shufflenet'
flops, params = profile(PretrainModel(network, 1, 10), inputs = (input, ))
flops_arr.append(flops)
params_arr.append(params)

for i in range(len(params_arr)):
    print('{} --- FLOPs: {} M, # Params: {} M'.format(model_arr[i], flops_arr[i]/1000**2, format(params_arr[i]/1000**2)))
```
